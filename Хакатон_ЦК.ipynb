```python
import pandas as pd
import numpy as np
```

Загрузка данных


```python
train=pd.read_csv("train.csv", delimiter=";")
test=pd.read_csv("test.csv", delimiter=";")
```

```python
train
```

```python
d=train.drop(columns=["id", "text"]).columns
d
```

Приведение данных к формату аналогичному примеру посылки для удобства работы


```python
all=[]
for i in range(len(train)):
  s=[]
  for column in d:
    #print(train[column][i], np.nan)
    if train[column][i]==1:
      s.append(0)
      s.append(0)
      s.append(1)
    elif train[column][i]==-1:
      s.append(1)
      s.append(0)
      s.append(0)
    elif train[column][i]==0:
      s.append(0)
      s.append(1)
      s.append(0)
    else:
      s.append(0)
      s.append(0)
      s.append(0)
  all.append(s)
all=pd.DataFrame(all)
all
```

```python
cols=[]
for i in d:
  cols.append(i+"_n")
  cols.append(i+"_0")
  cols.append(i+"_p")
```

```python
all.columns=cols
all
```

Векторизация с помощью TF-IDF

```python
all_data=pd.concat([train, test])
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(max_features=1000)
matrix = tfidf.fit_transform(all_data["text"]).toarray()
```

```python
from sklearn.multioutput import MultiOutputClassifier
from sklearn.linear_model import LogisticRegression


clf = MultiOutputClassifier(CatBoostClassifier()).fit(matrix[:len(train)], all)
```
